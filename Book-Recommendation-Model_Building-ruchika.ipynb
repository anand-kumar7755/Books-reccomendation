```python
# Pivot Table when index are book title and columns are user-id
# our aim is to find similarity between books
```

```python
pivot_table = country_users_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pivot_table
```

```python
# 706 books are rated by minimum 50 peoples
# and 810 users who rated more than 200 books
```

```python
pivot_table.fillna(0, inplace=True)
pivot_table
```

```python
# we will find similarities between books using cosine similarities
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score=cosine_similarity(pivot_table)
similarity_score
```

```python
# This is a 706 by 706 matrix which is showing euclidian distance between the 706 books
```

```python
similarity_score.shape
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    # index fetch
    index=np.where(pivot_table.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:6]
    
    for i in similar_items:
        print(pivot_table.index[i[0]])
```

```python
recommend("The Da Vinci Code")
```

```python
from scipy.sparse import csr_matrix
country_users_rating_pivot = country_users_rating.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating').fillna(0)
country_users_rating_matrix = csr_matrix(country_users_rating_pivot.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(country_users_rating_matrix)
```
